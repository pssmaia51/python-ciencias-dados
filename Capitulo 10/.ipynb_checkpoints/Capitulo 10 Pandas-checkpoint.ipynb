# Operadores Logicos para manupuliação de Dados com Pandas

```python
# Operador Logico AND
# Filtrar as vendas que ocorreram paa o segmento home office E para a região South
dsa_df[ (dsa_df.Segmento == 'Home Office') & (dsa_df.Regiao == 'South') ].head()
```

```python
# Operador Logico OU
# Filtrar as vendas que ocorreram paa o segmento home office OU para a região South
dsa_df[ (dsa_df.Segmento == 'Home Office') | (dsa_df.Regiao == 'South') ].tail()
```

```python
# Operador Logico NEGAÇÃO
# Filtrar as vendas que NÃO ocorreram paa o segmento home office e nem  para a região South
dsa_df[ (dsa_df.Segmento != 'Home Office') & (dsa_df.Regiao != 'South') ].sample(10)
```

```python
#Substituir um valor específico em uma string
import pandas as pd

# Criando um DataFrame
df = pd.DataFrame({'nomes': ['João Silva', 'Maria Silva', 'Pedro Silva']})
print(df)
```

```python
# Substituindo "Silva" por "Santos"
df['nomes'] = df['nomes'].str.replace('Silva', 'Santos')

print(df)
```

```python
#Remover Caracteres indesejados
df = pd.DataFrame({'valores': ['$1,000.00', '$500.00', '$3,250.75']})

print(df)
```

```python
# Removendo símbolos de dólar e vírgulas
df['valores'] = df['valores'].str.replace('[$,]', '', regex=True)
print(df)
```

```python
# Substituir partes especificas de uma string 
df = pd.DataFrame({'códigos': ['ABC123', 'XYZ456', 'DEF789']})
print(df)
```

```python
# Substituir letras por "XXX"
df['códigos'] = df['códigos'].str.replace(r'[A-Z]+', 'XXX', regex=True)

print(df)
```

```python
# Substituir valores especificos em varias condições 
df = pd.DataFrame({'status:': ['aprovado', 'pendente', 'reprovado', 'aprovado']})
print(df)
```

```python
# Substituindo status
df['status:'] = df['status:'].replace({'aprovado': 'OK', 'reprovado': 'NOK'})

print(df)
```

```python
# Substituir números ou partes de strings específicas
df = pd.DataFrame({'produtos': ['Produto 1', 'Produto 2', 'Produto 3']})
print(df)
# Substituir os números por "X"
```

```python
df['produtos'] = df['produtos'].str.replace(r'\d', 'X', regex=True)
print(df)
```

```python
# Imagine que o DataFrame seja assim:
data = {
    'ID_Pedido': ['123', '456', '789'],
    'Segmento': ['A', 'B', 'C']
}

dsa_df = pd.DataFrame(data)
print(dsa_df)
```

```python
# Concatenando
dsa_df['Pedido_Segmento'] = dsa_df['ID_Pedido'].str.cat(dsa_df['Segmento'], sep='-')

print(dsa_df)
```

```python
# Combinação de duas colunas 
# Criando um DataFrame de exemplo
data = {
    'Produto': ['Notebook', 'Celular', 'Tablet'],
    'Categoria': ['Eletrônicos', 'Eletrônicos', 'Eletrônicos']
}

df = pd.DataFrame(data)
print(df)
```

```python
# Combinando as colunas 'Produto' e 'Categoria' com um separador '-'
df['Produto_Categoria'] = df['Produto'] + ' - ' + df['Categoria']

print(df)
```

```python
# Concatenar múltiplas colunas com Pandas (str.cat)
data = {
    'Nome': ['João', 'Maria', 'Pedro'],
    'Sobrenome': ['Silva', 'Oliveira', 'Costa'],
    'Cidade': ['São Paulo', 'Rio de Janeiro', 'Curitiba']
}

df = pd.DataFrame(data)

print(df)
```

```python
# Concatenando 'Nome', 'Sobrenome' e 'Cidade' com separadores diferentes
df['Resumo'] = df['Sobrenome'].str.cat(df[['Nome', 'Cidade']], sep=', ')

print(df)
```

```python
# Adicionar texto adicional durante a combinação
data = {
    'Produto': ['Notebook', 'Celular', 'Tablet'],
    'Preco': [2500, 1500, 1200]
}

df = pd.DataFrame(data)
print(df)
```

```python
# Criando uma coluna que combina com texto adicional
df['Descricao'] = 'O ' + df['Produto'] + ' custa R$' + df['Preco'].astype(str)

print(df)
```

```python
# Graficos com Pandas
#Carrenado arquivo csv 
import pandas as pd

# Carregar o arquivo CSV no DataFrame
dsa_df = pd.read_csv('meudataset.csv')

# Visualizar os primeiros dados para garantir que foi carregado corretamente
print(dsa_df.head())
```

```python
import matplotlib.pyplot as plt

# Total de vendas por categoria
vendas_categoria = dsa_df.groupby('Categoria')['Valor_Venda'].sum()

# Definir uma lista de cores mais sóbrias, mas vivas
cores_sobrias_vivas = ['#1F3A57', '#4B9E6B', '#D96459', '#F1C232', '#6D6D6D']  # Azul marinho, verde musgo, vinho, mostarda e cinza escuro

# Criar o gráfico com as cores personalizadas
ax = vendas_categoria.plot(kind='bar', color=cores_sobrias_vivas[:len(vendas_categoria)], title='Vendas por Categoria')

# Adicionar rótulos nas barras
for i, v in enumerate(vendas_categoria):
    ax.text(i, v + 10, f'{v:.2f}', ha='center', va='bottom', color='black')

# Adicionar rótulos
plt.xlabel('Categoria')
plt.ylabel('Valor Total de Vendas')
plt.show()
```

```python
# Total de vendas por região
vendas_regiao = dsa_df.groupby('Regiao')['Valor_Venda'].sum()

# Criar o gráfico de pizza
vendas_regiao.plot(kind='pie', autopct='%1.1f%%', startangle=90, title='Proporção de Vendas por Região', colors=['gold', 'lightcoral', 'skyblue'])

# Remover rótulo desnecessário do eixo
plt.ylabel('')
plt.show()
```

```python
# Converter 'Data_Pedido' para datetime, se não for
dsa_df['Data_Pedido'] = pd.to_datetime(dsa_df['Data_Pedido'])

# Extrair o ano
dsa_df['Ano'] = dsa_df['Data_Pedido'].dt.year

# Agora, tente agrupar por 'Ano'
vendas_ano = dsa_df.groupby('Ano')['Valor_Venda'].sum()

# Criar o gráfico
vendas_ano.plot(kind='line', marker='o', title='Evolução de Vendas por Ano', color='green')

# Adicionar rótulos
plt.xlabel('Ano')
plt.ylabel('Valor Total de Vendas')
plt.grid(True)
plt.show()
```

```python
# Criar o gráfico de dispersão
dsa_df.plot.scatter(x='Quantidade', y='Valor_Venda', color='purple', title='Valor de Venda vs Quantidade')

# Adicionar rótulos
plt.xlabel('Quantidade')
plt.ylabel('Valor de Venda')
plt.show()
```

```python
# Criar o histograma
dsa_df['Valor_Venda'].plot(kind='hist', bins=10, color='orange', title='Distribuição dos Valores de Venda')

# Adicionar rótulos
plt.xlabel('Valor de Venda')
plt.ylabel('Frequência')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris

# Carregar o dataset Iris
iris = load_iris()

# Criar o DataFrame
iris_df = pd.DataFrame(data=iris.data, columns=iris.feature_names)

# Adicionar a coluna 'especies' com as classes das espécies
iris_df['especies'] = iris.target

# Mapear o target para os nomes das espécies
species_map = {0: 'Iris-setosa', 1: 'Iris-versicolor', 2: 'Iris-virginica'}
iris_df['especies'] = iris_df['especies'].map(species_map)

# Calcular a média das colunas agrupada por 'especies'
mean_values = iris_df.groupby('especies').mean()

# Criar o gráfico de barras
ax = mean_values.plot(kind='bar', figsize=(10, 6), colormap='viridis')

# Adicionar título e rótulos
plt.title('Média das Características por Espécie')
plt.xlabel('Espécies')
plt.ylabel('Valor Médio (cm)')

# Adicionar os valores dentro das barras
for p in ax.patches:
    ax.annotate(f'{p.get_height():.2f}', 
                (p.get_x() + p.get_width() / 2., p.get_height()), 
                xytext=(0, 5), 
                textcoords='offset points', 
                ha='center', va='bottom', color='black', fontweight='bold')

# Mostrar o gráfico
plt.show()

```

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris

# Carregar o dataset Iris
iris = load_iris()

# Criar o DataFrame
iris_df = pd.DataFrame(data=iris.data, columns=iris.feature_names)

# Adicionar a coluna 'especies' com as classes das espécies
species_map = {0: 'Iris-setosa', 1: 'Iris-versicolor', 2: 'Iris-virginica'}
iris_df['especies'] = iris.target
iris_df['especies'] = iris_df['especies'].map(species_map)

# Contar as ocorrências de cada espécie
species_counts = iris_df['especies'].value_counts()

# Criar o gráfico de pizza
plt.figure(figsize=(8, 6))
plt.pie(species_counts, labels=species_counts.index, autopct='%1.1f%%', startangle=90, colors=['#66b3ff','#99ff99','#ffcc99'])

# Adicionar título
plt.title('Distribuição das Espécies no Dataset Iris')

# Exibir o gráfico
plt.show()

```

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris

# Carregar o dataset Iris
iris = load_iris()

# Criar o DataFrame
iris_df = pd.DataFrame(data=iris.data, columns=iris.feature_names)

# Gerar KDE para cada variável com subplots
iris_df.plot.kde(subplots=True, figsize=(8, 8), layout=(2, 2), legend=False)

# Adicionar título geral à figura
plt.suptitle('Distribuição KDE das Variáveis do Dataset Iris', fontsize=16)

# Ajustar o layout
plt.tight_layout()

# Exibir o gráfico
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_iris

# Carregar o dataset Iris
iris = load_iris()

# Criar o DataFrame
iris_df = pd.DataFrame(data=iris.data, columns=iris.feature_names)

# Configuração para os gráficos
sns.set(style="whitegrid")

# Criar um gráfico KDE para cada variável no DataFrame
plt.figure(figsize=(12, 10))

for i, column in enumerate(iris_df.columns):
    plt.subplot(2, 2, i + 1)  # Ajuste para uma grid de 2x2
    sns.kdeplot(iris_df[column], fill=True, color="skyblue")  # Atualização: usar fill=True
    plt.title(f'KDE - {column}')

# Adicionar título geral à figura
plt.suptitle('Distribuição KDE das Variáveis do Dataset Iris', fontsize=16)

# Ajuste de layout para não sobrepor os gráficos
plt.tight_layout()

# Exibir o gráfico
plt.show()

```
