## Pergunta de Negócio 5:

### Qual Segmento Teve o Maior Total de Vendas?

Demonstre o resultado através de um gráfico de pizza.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Agrupar por Segmento e somar o Valor_Venda
vendas_por_segmento = df_dsa.groupby('Segmento')['Valor_Venda'].sum().reset_index().sort_values(by = 'Valor_Venda',
                                                                                      ascending = False)

# Ordenar corretamente (sem especificar 'by', pois é uma Série)


vendas_por_segmento.head(10)
```

```python
import matplotlib.pyplot as plt

# Função para converter os dados em valor absoluto
def autopct_format(values): 
    def my_format(pct): 
        total = sum(values) 
        val = int(round(pct * total / 100.0))
        return f'$ {val:d}'
    return my_format
```

```python
# Tamanho da figura
plt.figure(figsize=(10, 10))

# Gráfico de pizza corrigido
plt.pie(vendas_por_segmento['Valor_Venda'], 
        labels=vendas_por_segmento['Segmento'], 
        autopct=autopct_format(vendas_por_segmento['Valor_Venda']), 
        startangle=90, 
        colors=['lightblue', 'lightcoral', 'lightgreen', 'gold', 'lightpink'])

# Adicionando um círculo central para um efeito "donut"
centre_circle = plt.Circle((0, 0), 0.75, fc='white')
fig = plt.gcf()
fig.gca().add_artist(centre_circle)

# Título e anotação
plt.annotate(text=f'Total de Vendas: $ {int(vendas_por_segmento["Valor_Venda"].sum())}', xy=(-0.2, 0))
plt.title('Total de Vendas Por Segmento')
plt.show()
```

```python
import matplotlib.pyplot as plt

# Criar gráfico de pizza
plt.figure(figsize=(8, 8))

# Definir explode para destacar o maior segmento
explode = (0.1, 0, 0)  # Apenas o primeiro segmento será destacado

plt.pie(vendas_por_segmento['Valor_Venda'], 
        labels=vendas_por_segmento['Segmento'], 
        autopct='%1.1f%%',
        startangle=90, 
        shadow=True, 
        explode=explode,  # Explode adicionado
        colors=['lightblue', 'lightcoral', 'lightgreen'])

plt.title("Total de Vendas por Segmento")
plt.show()

```

## Pergunta de Negócio 8 (Desafio Nível Master):

### Considere Que a Empresa Decida Conceder o Desconto de 15% do Item Anterior. Qual Seria a Média do Valor de Venda Antes e Depois do Desconto?

```python
# Aplicar o desconto de 15% nas vendas que são maiores que 1000
df_dsa['Valor_Venda_Com_Desconto'] = df_dsa['Valor_Venda'].apply(
    lambda x: x * 0.85 if x > 1000 else x)

# Calcular a média do Valor de Venda antes e depois do desconto
media_antes_desconto = df_dsa['Valor_Venda'].mean()
media_depois_desconto = df_dsa['Valor_Venda_Com_Desconto'].mean()

# Exibir o resultado
print(f"Média do Valor de Venda Antes do Desconto: {media_antes_desconto:.2f}")
print(f"Média do Valor de Venda Depois do Desconto: {media_depois_desconto:.2f}")
```

```python
# Alternativa 2
df_dsa['Valor_Venda_Desconto'] = df_dsa['Valor_Venda'] - (df_dsa['Valor_Venda'] * df_dsa['Desconto'])

# Filtrando as vendas antes do desconto de 15%
df_dsa_p8_vendas_antes_desconto = df_dsa.loc[df_dsa['Desconto'] == 0.15, 'Valor_Venda']

# Filtrando as vendas depois do desconto de 15%
df_dsa_p8_vendas_depois_desconto = df_dsa.loc[df_dsa['Desconto'] == 0.15, 'Valor_Venda_Desconto']

# Calcula a média das vendas antes do desconto de 15%
media_vendas_antes_desconto = df_dsa_p8_vendas_antes_desconto.mean()

# Calcula a média das vendas depois do desconto de 15%
media_vendas_depois_desconto = df_dsa_p8_vendas_depois_desconto.mean()

print(f"Média do Valor de Venda Antes do Desconto: {media_antes_desconto:.2f}")
print(f"Média do Valor de Venda Depois do Desconto: {media_depois_desconto:.2f}")
```

## Pergunta de Negócio 10 (Desafio Nível Master Ninja das Galáxias):

### Qual o Total de Vendas Por Categoria e SubCategoria, Considerando Somente as Top 12 SubCategorias? 

Demonstre tudo através de um único gráfico.

```python
# Minha solução
import pandas as pd
import matplotlib.pyplot as plt

# Agrupar por Categoria e Subcategoria e somar as vendas
total_vendas_categoria_subcategoria = df_dsa.groupby(['Categoria', 'SubCategoria'])['Valor_Venda'].sum().reset_index()

# Calcular o total de vendas por SubCategoria e pegar as top 12
top_subcategorias = total_vendas_categoria_subcategoria.groupby('SubCategoria')['Valor_Venda'].sum().sort_values(ascending=False).head(12).index

# Filtrar os dados para considerar apenas as top 12 Subcategorias
top_vendas_categoria_subcategoria = total_vendas_categoria_subcategoria[total_vendas_categoria_subcategoria['SubCategoria'].isin(top_subcategorias)]

# Agrupar por Categoria e SubCategoria novamente
total_vendas_categoria_subcategoria_top = top_vendas_categoria_subcategoria.groupby(['Categoria', 'SubCategoria'])['Valor_Venda'].sum().unstack().fillna(0)

# Gerar gráfico de barras empilhadas
ax = total_vendas_categoria_subcategoria_top.plot(kind='bar', stacked=True, figsize=(12, 8), cmap='Set2')

# Adicionar título e rótulos
plt.title("Total de Vendas por Categoria e SubCategoria (Top 12 Subcategorias)", fontsize=16)
plt.xlabel("Categoria", fontsize=12)
plt.ylabel("Total de Vendas", fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.legend(title='SubCategoria', bbox_to_anchor=(1.05, 1), loc='upper left')

# Adicionar os valores de vendas nas barras
for p in ax.patches:
    height = p.get_height()
    width = p.get_width()
    x = p.get_x() + width / 2
    y = p.get_y() + height / 2
    ax.text(x, y, f'{height:.2f}', ha='center', va='center')

# Ajuste de layout
plt.tight_layout()
plt.show()


```
