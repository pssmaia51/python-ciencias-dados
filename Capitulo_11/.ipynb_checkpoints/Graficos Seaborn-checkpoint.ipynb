```python
import seaborn as sns

# Carregar o dataset 'tips'
dados = sns.load_dataset("tips")

# Exibir as primeiras linhas do dataset
print(dados.head())
```

```python
# Criar jointplot com regressão linear
sns.jointplot(data=dados, x="total_bill", y="tip", kind="reg")

# Exibir o gráfico
plt.show()
```

```python
# Criar gráfico de regressão separando por fumantes e não fumantes
sns.lmplot(data=dados, x="total_bill", y="tip", col="smoker")

# Exibir o gráfico
plt.show()
```

```python
import pandas as pd
import random
import seaborn as sns
import matplotlib.pyplot as plt

# Criando um DataFrame vazio
df = pd.DataFrame()

# Adicionando dados aleatórios para 'Idade' e 'Peso'
df['Idade'] = random.sample(range(20, 100), 30)
df['Peso'] = random.sample(range(55, 150), 30)

# Verificando o tamanho do DataFrame
print(df.shape)

# Exibindo as primeiras linhas do DataFrame
print(df.head())

# Criando o gráfico de regressão linear
sns.lmplot(data=df, x="Idade", y="Peso", fit_reg=True)

# Exibir o gráfico
plt.show()
```

```python
import pandas as pd
import random
import seaborn as sns
import matplotlib.pyplot as plt

# Criando um DataFrame vazio
df = pd.DataFrame()

# Adicionando dados aleatórios para 'Idade' e 'Peso'
df['Idade'] = random.sample(range(20, 100), 30)
df['Peso'] = random.sample(range(55, 150), 30)

# Exibir as primeiras linhas do DataFrame
print(df.head())

# Criando gráfico de densidade para 'Idade'
sns.kdeplot(data=df, x="Idade", fill=True, color="blue", linewidth=2)

# Criando gráfico de densidade para 'Peso'
sns.kdeplot(data=df, x="Peso", fill=True, color="red", linewidth=2)

# Exibir o gráfico
plt.show()
```

```python
import numpy as np
import pandas as pd

np.random.seed(42)

n = 1000

pct_smokers = 0.2

# Variáveis
flag_fumante = np.random.rand(n) < pct_smokers

idade = np.random.normal(40, 10, n)  # Corrigido: 'normal' ao invés de 'norma1'

altura = np.random.normal(170, 10, n)

peso = np.random.normal(70, 10, n)

# DataFrame
dados = pd.DataFrame({
    'altura': altura,
    'peso': peso,
    'flag_fumante': flag_fumante
})

# Mapeando flag_fumante para valores de texto
dados['flag_fumante'] = dados['flag_fumante'].map({True: 'Fumante', False: 'Não Fumante'})

# Exibindo as primeiras linhas do DataFrame
print(dados.head())
```

```python
dados.shape
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Definindo o estilo
sns.set(style='ticks')

# Criando o gráfico de dispersão
sns.scatterplot(x='altura',
                y='peso',
                data=dados,
                hue='flag_fumante',
                palette=['tab:blue', 'tab:orange'],
                s=100)  # Tamanho dos pontos ajustado

# Ajustando os rótulos e título
plt.xlabel('Altura (cm)')
plt.ylabel('Peso (kg)')
plt.title('Relação entre altura e peso de fumantes e não fumantes')

# Removendo as bordas
sns.despine()

# Exibindo o gráfico
plt.show()
```
