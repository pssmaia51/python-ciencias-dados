## Try, Execpt, Finally

```python
try:
    8 + 'a'
except TypeError:
    print('Operação invalida, não e possivel somar numero com letra')
```

```python
#Utilizando Try, except else
try:
    f = open('C:/PythonDSA/Capitulo 06/arquivos/testandoerros.txt', 'w')
    f.write('Gravando no arquivo')
except IOError:
    print('Erro: arquivo não encontrado ou bloquedo para gravação')
else:
    print('Dados gravados com sucesso no arquivo destino')
    f.close()
```

```python
#Utilizando Try, except else
try:
    f = open('C:/PythonDSA/Capitulo 06/arquivos/testandoerros.txt', 'r')
except IOError:
        print('Erro: arquivo não encontrado ou não pode ser lido')
else:
    print('Dados gravados com sucesso no arquivo destino')
    f.close()
```

```python
#Utilizando Try, except else e finally
try:
    f = open('C:/PythonDSA/Capitulo 06/arquivos/testandoerros.txt', 'w')
    f.write('Gravando no arquivo')
except IOError:
    print('Erro: arquivo não encontrado ou bloquedo para gravação')
else:
    print('Dados gravados com sucesso no arquivo destino')
    f.close()
finally:
    print('Mostrando que os comandos no bloco finally são sempre execultados')
```

```python
def askint():
    try:
        val = int(input("Digite um numero: "))
    except: 
        print("Voce não digitou nem um numero")
    finally:
        print("Obrigado")

askint()    
```

```python
def askint():
    try:
        val = int(input("Digite um número: "))
    except ValueError:
        print("Você não digitou um número.")
        val = int(input("Tente novamente, digite um número: "))
    finally:
        print("Obrigado")
    print(val)

askint()
        
```

```python
def askint():
    while True:
        try:
            val = int(input("Digite um número: "))
        except ValueError:
            print("Você não digitou um número.")
            continue
        else:
            print("Obrigado por entrar com um número.")
            break
        finally:
            print("Fim da Execução")
    
    print(val)

askint()
            
```
