### Pacotes e modulos
#### Em Python um modulo e um arquivo (Script) que contem codigo python e pode ser importado para em outros arquivos python. Ele
### e usado para compartilhar funções, classes e variaveis entre arquivos.
### Já um pacote e uma coleção de modulos organizados em uma estrutura de diretorio. Ele permite a divisão de uma arquivo em 
### multiplos modolos o que facilita a manutenção e o desenvolvimento.

```python
# Usando metodo numpy sqrt para calcular a raiz quadrada de cada elemento em um array
import numpy as np
# Criando um array 1D
array = np.array([1, 2, 3, 4])

# Operações matemáticas vetorizadas
array_soma = array + 5  # [6, 7, 8, 9]
array_produto = array * 2  # [2, 4, 6, 8]

# Criando um array 2D (matriz)
matriz = np.array([[1, 2], [3, 4]])

# Multiplicação de matrizes
matriz_resultado = np.dot(matriz, matriz)
```

```python
# Importando pacote random usada para gerar numeros aletorios
import random 
```

```python
# Usando metodo choice (escolher)
random.choices(['Abacate', 'Banana', 'Laranja'])

# Lista de opções
opcoes = ['maçã', 'banana', 'cereja']

# Escolhendo um item aleatório
print(random.choice(opcoes))  # Ex: 'banana'

# Escolhendo 2 itens aleatórios com reposição
print(random.choices(opcoes, k=2))  # Ex: ['maçã', 'maçã']

# Escolhendo 2 itens aleatórios sem reposição
print(random.sample(opcoes, k=2))  # Ex: ['cereja', 'banana']
```

```python
# Usando metodo sample (amostra)
random.sample(range(100), 10)
```

```python
# Embaralhar uma sequencia
cartas = ['A', 'B', 'C', 'D']

random.shuffle(cartas)
print(cartas)  # Ex: ['C', 'A', 'D', 'B']
```

```python
# Número flutuante entre 0.0 e 1.0
print(random.random())  # Ex: 0.3745

# Número inteiro entre 1 e 10
print(random.randint(1, 10))  # Ex: 7

# Número flutuante entre 1.5 e 10.5
print(random.uniform(1.5, 10.5))  # Ex: 8.234
```

```python
# Gerar Números Aleatórios com Distribuições Específicas:
# Número aleatório de uma distribuição normal
print(random.gauss(0, 1))  # Ex: -0.9786

# Número aleatório de uma distribuição exponencial
print(random.expovariate(1.5))  # Ex: 0.4385
```

```python
# Importando pacote statistics usada para gerar numeros aletorios
import statistics
```

```python
#  Cálculo de Medidas de Tendência Central:
data = [1, 2, 2, 3, 4, 5, 5, 5, 6]

# Média
print(statistics.mean(data))  # Ex: 3.6666666666666665

# Mediana
print(statistics.median(data))  # Ex: 4

# Moda
print(statistics.mode(data))  # Ex: 5
```

```python
# Cálculo de Medidas de Dispersão:
# Variância
print(statistics.variance(data))  # Ex: 3.25

# Desvio padrão
print(statistics.stdev(data))  # Ex: 1.8027756377319946
```

```python
# Cálculo de Medidas de Posição:
# Quantis (quartis, por exemplo)
print(statistics.quantiles(data, n=4))  # Ex: [2, 4, 5]
```

```python
# Importando o modulo request do pacote urllib usado para trazer url's para dentro do nosso ambiente Python
import urllib.request

# Especifica url
url = 'https://www.python.org/'

# Abra o URL e busque a resposta
response = urllib.request.urlopen(url)

# Leia o conteúdo da página
web_content = response.read()

# Imprimir o conteudo da pagina (opcional)
print(web_content)
```

```python
# Variavel resposta armazena o objeto de conexão à url passada como parametro
resposta = urllib.request.urlopen('http://python.org/')
```
